The data explains the various job designation in data science and related salaries from across the world. \

*Approach*: \

* Visualization of the data to draw conclusions using different types of plots to visualize the categorical data vs the numerical data

# Bivariate analysis

```python
Year_mean = data.groupby('work_year')['salary_in_usd'].mean()
Year_std = data.groupby('work_year')['salary_in_usd'].std()
print(' Mean Salary in USD per year')
print(Year_mean)
print(' STD Salary in USD per year')
print(Year_std)
print(' Max. Salary in USD per year')
print(data.groupby('work_year')['salary_in_usd'].max())
print(' Min Salary in USD per year')
print(data.groupby('work_year')['salary_in_usd'].min())
```

```python
Job_title_mean = data.groupby('job_title')['salary_in_usd'].mean()
Job_title_std = data.groupby('job_title')['salary_in_usd'].std()
print(Job_title_mean.nlargest(10))
print(Job_title_mean.nsmallest(10))
```

```python
Job_category_mean = data.groupby('job_category')['salary_in_usd'].mean()
Job_category_std = data.groupby('job_category')['salary_in_usd'].std()
print(' Mean Salary in USD per Job Category')
print(Job_category_mean)
print(' STD Salary in USD per Job Category')
print(Job_category_std)
print('Max Salary per Job Category')
print(data.groupby('job_category')['salary_in_usd'].max())
print('Min Salary per Job Category')
print(data.groupby('job_category')['salary_in_usd'].min())
```

```python
Employee_residence_mean = data.groupby('employee_residence')['salary_in_usd'].mean()
Employee_residence_std = data.groupby('employee_residence')['salary_in_usd'].std()
```

```python
Experience_level_mean = data.groupby('experience_level')['salary_in_usd'].mean()
Experience_level_std = data.groupby('experience_level')['salary_in_usd'].std()
print('Mean Salary in USD per Experience level' )
print(Experience_level_mean)
print('STD Salary in USD per Experience level')
print(Experience_level_std)
print(data.groupby('experience_level')['salary_in_usd'].max())
print(data.groupby('experience_level')['salary_in_usd'].min())
```
